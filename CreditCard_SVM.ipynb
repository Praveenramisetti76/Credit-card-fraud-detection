```python
X = df.drop('Class', axis=1)   # features
y = df['Class']                # target
```

```python
from sklearn.model_selection import train_test_split

# Identify and remove rows where 'y' (Class) is NaN
# This is crucial because stratify=y cannot handle NaN values
nan_indices = y[y.isna()].index
X_cleaned = X.drop(nan_indices)
y_cleaned = y.drop(nan_indices)

X_train, X_test, y_train, y_test = train_test_split(
    X_cleaned, y_cleaned, test_size=0.2, random_state=42, stratify=y_cleaned
)

print(X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Scale columns 0 and 29 by their integer positions using .iloc
X_train.iloc[:, [0, 29]] = scaler.fit_transform(X_train.iloc[:, [0, 29]])
X_test.iloc[:, [0, 29]] = scaler.transform(X_test.iloc[:, [0, 29]])
```

```python
from sklearn.svm import LinearSVC

svm_clf = LinearSVC(class_weight='balanced')
svm_clf.fit(X_train, y_train)
```

```python
y_pred = svm_clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
```

```python
from sklearn.metrics import roc_auc_score

y_scores = svm_clf.decision_function(X_test)
roc = roc_auc_score(y_test, y_scores)
print("ROC-AUC:", roc)
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

# Generate FPR, TPR for ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_scores)

# Plot the ROC Curve
plt.figure(figsize=(7, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc:.4f})")
plt.plot([0, 1], [0, 1], linestyle="--", color="grey")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report

# Train
svm_clf = LinearSVC(class_weight='balanced')
svm_clf.fit(X_train, y_train)

# Predict
y_pred = svm_clf.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Extra
print(classification_report(y_test, y_pred))
```
